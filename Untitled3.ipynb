```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the iris dataset from seaborn library
df=sns.load_dataset('iris')
```

```python
# using the head() function to take a look at the first few rows of the dataset
df.head()
```

```python
# using the shape attribute to get the dimensions of the dataset
df.shape
```

```python
df.columns
```

```python
# using the describe() function to get a statistical overview of the dataset

df.describe()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
```

```python
sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=df)
```

```python
sns.scatterplot(x='sepal_width', y='petal_width', hue='species', data=df)
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
```

```python
numerical_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
for col in numerical_cols:
    plt.figure(figsize=(8, 5))
    plt.hist(df[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[numerical_cols])
plt.title('Box Plots of Iris Features')
plt.ylabel('cm')
plt.show()
```
